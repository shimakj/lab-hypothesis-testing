**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# Filter data for Dragon and Grass types
dragon_hp = df[df['Type 1'] == 'Dragon']['HP'].dropna()
grass_hp = df[df['Type 1'] == 'Grass']['HP'].dropna()

# Check for normality using Shapiro-Wilk test
shapiro_dragon = st.shapiro(dragon_hp)
shapiro_grass = st.shapiro(grass_hp)

# Check for variance homogeneity using Levene's test
levene_test = st.levene(dragon_hp, grass_hp)

# Perform independent t-test
t_test = st.ttest_ind(dragon_hp, grass_hp, equal_var=levene_test.pvalue > 0.05)

# Compile results
results = {
    "Shapiro-Wilk Dragon": shapiro_dragon,
    "Shapiro-Wilk Grass": shapiro_grass,
    "Levene's Test": levene_test,
    "T-Test": t_test
}

results
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Define columns to test
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Split the data
legendary_stats = df[df['Legendary'] == True][stats_columns]
non_legendary_stats = df[df['Legendary'] == False][stats_columns]

# Initialize a dictionary to store the results
results = {}

# Loop through each stat column and perform T-Test
for stat in stats_columns:
    # Shapiro-Wilk Test for normality
    shapiro_legendary = st.shapiro(legendary_stats[stat].dropna())
    shapiro_non_legendary = st.shapiro(non_legendary_stats[stat].dropna())
    
    # Levene's Test for equal variances
    levene_test = st.levene(legendary_stats[stat].dropna(), non_legendary_stats[stat].dropna())
    
    # T-Test (adjust for equal/unequal variances based on Levene's result)
    equal_var = levene_test.pvalue > 0.05
    t_test = st.ttest_ind(legendary_stats[stat].dropna(), non_legendary_stats[stat].dropna(), equal_var=equal_var)
    
    # Store results
    results[stat] = {
        "Shapiro-Wilk Legendary": shapiro_legendary.pvalue,
        "Shapiro-Wilk Non-Legendary": shapiro_non_legendary.pvalue,
        "Levene's Test": levene_test.pvalue,
        "T-Test Statistic": t_test.statistic,
        "T-Test p-value": t_test.pvalue,
        "Equal Variance Assumption": equal_var
    }

print(results)
```

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# School and Hospital coordinates
school_coords = (-118, 34)
hospital_coords = (-122, 37)
```

```python

# Function to calculate E-distance
def euclidean_distance(coord1, coord2):
    return np.sqrt((coord1[0] - coord2[0]) ** 2 + (coord1[1] - coord2[1]) ** 2)
```

```python

# Calculate distances
df['Distance_School'] = df.apply(lambda row: euclidean_distance((row['longitude'], row['latitude']), school_coords), axis=1)
df['Distance_Hospital'] = df.apply(lambda row: euclidean_distance((row['longitude'], row['latitude']), hospital_coords), axis=1)
```

```python

# Classify as 'Close' or 'Far'
df['Proximity'] = np.where((df['Distance_School'] < 0.5) | (df['Distance_Hospital'] < 0.5), 'Close', 'Far')
```

```python

# Split data
close_prices = df[df['Proximity'] == 'Close']['median_house_value']
far_prices = df[df['Proximity'] == 'Far']['median_house_value']
```

```python


# Normality Test
shapiro_close = st.shapiro(close_prices.dropna())
shapiro_far = st.shapiro(far_prices.dropna())
```

```python


# Variance Homogeneity Test
levene_test = st.levene(close_prices.dropna(), far_prices.dropna())
```

```python


# T-Test
equal_var = levene_test.pvalue > 0.05
t_test = st.ttest_ind(close_prices.dropna(), far_prices.dropna(), equal_var=equal_var)
```

```python


# Compile results
results = {
    "Shapiro-Wilk Close": shapiro_close.pvalue,
    "Shapiro-Wilk Far": shapiro_far.pvalue,
    "Levene's Test": levene_test.pvalue,
    "T-Test Statistic": t_test.statistic,
    "T-Test p-value": t_test.pvalue,
    "Equal Variance Assumption": equal_var
}

# Display the results
print(pd.DataFrame([results]))
```
